```python
import pandas as pd
import sqlite3

conn = sqlite3.connect('projeto_grupo5.db')
cursor = conn.cursor()

cursor.execute('DELETE FROM "Transaction"')
cursor.execute('DELETE FROM "Officer"')
cursor.execute('DELETE FROM "Project"')
cursor.execute('DELETE FROM "Agency"')
conn.commit()

df = pd.read_csv('g65_Agencies_Projects.csv', sep=';')

df['amount'] = df['amount'].astype(str).str.replace(',', '.').astype(float)

print("Gavetas limpas! A começar a arrumação...")

agencies = df[['agency_id', 'name']].dropna().drop_duplicates(subset=['agency_id'])
agencies.columns = ['id', 'name']
agencies.to_sql('Agency', conn, if_exists='append', index=False)
print("✔ Agências arrumadas!")

projects = df[['project_id', 'project_name', 'category', 'started_date']].dropna().drop_duplicates(subset=['project_id'])
projects.columns = ['id', 'name', 'category', 'started_date']
projects.to_sql('Project', conn, if_exists='append', index=False)
print("✔ Projetos arrumados!")

officers = df[['officer_id', 'role', 'agency_id']].dropna(subset=['officer_id']).drop_duplicates(subset=['officer_id'])
officers['officer_id'] = officers['officer_id'].astype(str).str.replace(',', '.').astype(float).astype(int)
officers.columns = ['id', 'role', 'agency_id']
officers.to_sql('Officer', conn, if_exists='append', index=False)
print("✔ Oficiais arrumados!")

transactions = df[['agency_id', 'projbect_id', 'payment_date', 'amount']].dropna()
transactions = transactions.reset_index(drop=True)
transactions['id'] = transactions.index + 1
transactions = transactions[['id', 'agency_id', 'project_id', 'payment_date', 'amount']]
transactions.to_sql('Transaction', conn, if_exists='append', index=False)
print("✔ Transações arrumadas!")

conn.close()
print("🎉 TUDO PRONTO! O Arquivo está cheio e sem erros.")
```
